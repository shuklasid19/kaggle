```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x, y, 
                                                    test_size=0.2,
                                                    random_state=0)

print("Shape of x_train :", x_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of y_test :", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from scipy import stats
from scipy.stats import norm

plt.rcParams['figure.figsize'] = (10, 7)
sns.distplot(data['Chance of Admit '], fit=norm)

#getting 
mu ,sigma = norm.fit(data['Chance of Admit '])
plt.legend(["mu {:.2f} and sigma {:.2f}".format(mu, sigma)], 
           loc = 2)

plt.title("distribution of chances of admisttion")
plt.show()
```

```python

from scipy import stats

plt.rcParams['figure.figsize'] = (7, 5)
stats.probplot(data['Chance of Admit '], plot = plt)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
linear_pred = linreg.predict(x_test)
```

```python
mse = mean_squared_error(y_test, linear_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, linear_pred)
```

```python
rmse
```

```python
r2
```

```python
from sklearn.svm import SVR

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

svr =  SVR(kernel = "linear")
svr.fit(x_train  , y_train)

svr_pred = svr.predict(x_test)

mse = mean_squared_error(y_test, svr_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, svr_pred)

print("root mean squared error", rmse)
print("r squared error", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
rfr.fit(x_train , y_train)

rfr_pred = linreg.predict(x_test)
```

```python
mse = mean_squared_error(y_test, rfr_pred)
rmse = np.sqrt(mse)
r2=r2_score(y_test, rfr_pred)

print(rmse, r2)
```

```python
from xgboost.sklearn import XGBRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

xgb = XGBRegressor()
xgb.fit(x_train, y_train)

xgb_pred = xgb.predict(x_test)

mse = mean_squared_error(y_test, xgb_pred)
rmse = np.sqrt(mse)
r2 =  r2_score(y_test, xgb_pred)

print(rmse, r2)
```

```python
from sklearn.tree import ExtraTreeRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

etr = ExtraTreeRegressor()
etr.fit(x_train ,y_train)

etr_pred = etr.predict(x_test)

mse = mean_squared_error(y_test, etr_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, etr_pred)

print(rmse, r2)
```

```python
# stacked predictions

stacked_predictions = np.column_stack((linear_pred, svr_pred, rfr_pred))

# specifying the meta model
meta_model = LinearRegression()
meta_model.fit(stacked_predictions, y_test)

# final predictions
stacked_predictions = (meta_model.predict(stacked_predictions))

mse = mean_squared_error(y_test, stacked_predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, stacked_predictions)

print("Root Mean Squared Error : ",rmse)
print("R-Squared Error:", r2)
```
